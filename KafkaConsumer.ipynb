```python
import json
import boto3
from confluent_kafka import Consumer, KafkaException
from botocore.exceptions import NoCredentialsError

# Initialize Kafka Consumer
consumer = Consumer({
    'bootstrap.servers': '13.42.36.106:9092',  # IP address
    'group.id': 'my_placeholder_group',        
    'auto.offset.reset': 'earliest'            # Start reading from the earliest offset
})

# Subscribe to the topic
consumer.subscribe(['demo_test'])

# Initialize S3 client
s3 = boto3.client('s3')  # to make sure AWS credentials are configured

# Poll and process messages
try:
    count = 0
    while True:
        msg = consumer.poll(timeout=1.0)  # Poll for messages with a timeout
        if msg is None:
            continue  # No message received, continue polling

        if msg.error():
            raise KafkaException(msg.error())

        # error handling for non-JSON messages
        try:
            message_value = msg.value().decode('utf-8')
            if not message_value.strip():  # If message is empty or just whitespace
                print("Received an empty message, skipping...")
                continue

            # Converting the message to JSON
            message_data = json.loads(message_value)
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}. Message value: {message_value}")
            continue  # Skip this message if it fails to decode

        # your S3 bucket and file path
        bucket_name = 'kafka-stock-market-project-1'  # your S3 bucket name
        file_path = f"stock_market_{count}.json"

        # Uploading to S3
        try:
            s3.put_object(Bucket=bucket_name, Key=file_path, Body=json.dumps(message_data))
            print(f"Uploaded {file_path} to S3")
            count += 1
        except NoCredentialsError:
            print("S3 credentials not available. Check your AWS credentials configuration.")

finally:
    consumer.close()
```
